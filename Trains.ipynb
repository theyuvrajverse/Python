```python
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
ax = sns.countplot(x= 'Gender' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Purchase'].sum().sort_values(by='Purchase',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Purchase'].sum().sort_values(by='Purchase',ascending=False)
```

```python
sns.barplot(x='Gender',y='Purchase',data=sales_gen)
```

```python
sns.countplot(data=df,x='Age')
```

```python
ax = sns.countplot(x= 'Age' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
bx= sns.countplot(data=df, x='Age', hue='Gender')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
df.groupby(['Age'],as_index=False)['Purchase'].sum().sort_values (by='Purchase',ascending=False)
```

```python
# Purchase vs Age
sales_age=df.groupby(['Age'],as_index=False)['Purchase'].sum().sort_values (by='Purchase',ascending=False)
sns.barplot(x='Age',y='Purchase',data=sales_age)
```

```python
sale_city= df.groupby(['City_Category'],as_index =False)['Purchase'].sum().sort_values (by="Purchase",ascending=False)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='City_Category',y='Purchase',data=sale_city)
```

```python
cx = sns.countplot(data= df,x='Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
Marital_Purchase=df.groupby(['Marital_Status','Gender'],as_index=False)['Purchase'].sum().sort_values(by='Purchase',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Purchase',hue='Gender',data=Marital_Purchase)
```

```python
dx=sns.countplot(x='Occupation',data=df)
sns.set(rc={"figure.figsize":(22,5)})
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
Occupation_Purchase=df.groupby(['Occupation'],as_index=False)['Purchase'].sum().sort_values(by=['Purchase'],ascending=False)
sns.barplot(x='Occupation',y='Purchase',data=Occupation_Purchase)
```

```python
citywise_purchase=df.groupby(['Stay_In_Current_City_Years'],as_index=False)['Purchase'].sum().sort_values(by=["Purchase"],ascending=False)
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Stay_In_Current_City_Years',y='Purchase',data=citywise_purchase)
```

```python
ex=sns.countplot(x='Product_Category_1',data=df)
sns.set(rc={"figure.figsize":(30,10)})

for bars in ex.containers:
    ex.bar_label(bars)
```

```python
productwise_purchase=df.groupby(['Product_Category_1'],as_index=False)['Purchase'].sum().sort_values(by=["Purchase"],ascending=False)
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Product_Category_1',y='Purchase',data=productwise_purchase)
```

```python
most_purchased_productID=df.groupby(['Product_Category_1'],as_index=False)['Purchase'].sum().sort_values(by=['Purchase'],ascending=False)
sns.barplot(x='Product_Category_1',y='Purchase',data=most_purchased_productID)
```
